```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact, interact_manual, IntSlider, Select, Layout
from IPython.display import Image
%matplotlib inline
```

```python
def loadFile(depthPath, rgbPath, GTPath, npyPath):
    depth = plt.imread(depthPath)
    rgb = plt.imread(rgbPath)
    GT = plt.imread(GTPath)
    npyFile = np.load(npyPath)
    return depth, rgb, GT, npyFile
```

```python
def showImg(rgb, depth, npyFile, GT):
    fig = plt.figure(figsize=(18, 12))
    fig.add_subplot(1, 4, 1)
    plt.imshow(rgb)
    fig.add_subplot(1, 4, 2)
    plt.imshow(depth[:,:,1], cmap='gray')
    fig.add_subplot(1, 4, 3)
    plt.imshow(npyFile.reshape(512,512))
    fig.add_subplot(1, 4, 4)
    plt.imshow(GT)
    plt.show()
```

```python
idx = str(36)
rgbPath = f'./dataset/endoscopy/val/rgb/{idx}.jpg'
npyPath = f'./depthformer_swint_endoscopy_results/{idx}.npy'
depthPath = f'./depthformer_swint_endoscopy_results/{idx}.jpg'
GTPath = f'./dataset/endoscopy/val/depth/{idx}.jpg'
depth, rgb, GT, npyFile = loadFile(depthPath, rgbPath, GTPath, npyPath)
showImg(rgb, depth, npyFile, GT)
```
